```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('../../dtp_index/italy-dtp-'+str(2011)+'-'+str(6)+'-index.csv', header = 0, names=['days', 'time', 'lat', 'lon', 'discharge', 'temp', 'prec'])
#df = pd.read_csv('../../italy-dtp-2017-12.csv', header = 0, names=['days', 'time', 'lat', 'lon', 'discharge', 'temp', 'prec'])
df_drop = df.drop(columns = ['days', 'time', 'discharge', 'temp', 'prec'], inplace=False)
df['coords'] = df_drop.apply(tuple, axis=1)
df['time_2'] = df['time']
df = df.set_index(['coords', 'time'])
df.tail()
```
